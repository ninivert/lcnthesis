# Linear rate networks

Are the fixed points stable ?

Looks like $h = 0$ is stable, and $h = \xi_\mu$ is semistable (stability gets better as $N \rightarrow \infty$)

2023-06-05 correction : now we know the linear network's stability has eigenvalue zero, but numerically it seem at finite $N$ that the eigenvalue is slightly negative

```python
import matplotlib.pyplot as plt
import numpy as np
from neurodyn import *
```

## exclude self connections = False

self connections are present

```python
rnn = LowRankRNN.new_valentin(p=2, N=50_000, phi=identity, I_ext=zero, exclude_self_connections=False)
print(rnn)
res = rnn.simulate(h0=rnn.F[:, 0], t_span=(0, 1000), dt_max=5)
plot_neuron_trajectory(res);
```

```python
rnn = LowRankRNN.new_valentin(p=2, N=1_000_000, phi=identity, I_ext=zero, exclude_self_connections=False)
print(rnn)
res = rnn.simulate(h0=rnn.F[:, 0], t_span=(0, 1000), dt_max=5)
plot_neuron_trajectory(res);
```

```python
rnn = LowRankRNN.new_valentin(p=2, N=1_000_000, phi=identity, I_ext=zero, exclude_self_connections=False)
print(rnn)
res = rnn.simulate(h0=np.zeros(rnn.N), t_span=(0, 1000), dt_max=5)
plot_neuron_trajectory(res);
```

```python
rnn = LowRankRNN.new_valentin(p=2, N=1_000_000, phi=identity, I_ext=zero, exclude_self_connections=False)
print(rnn)
res = rnn.simulate(h0=np.random.default_rng(42).random(rnn.N), t_span=(0, 1000), dt_max=5)
plot_neuron_trajectory(res);
```
