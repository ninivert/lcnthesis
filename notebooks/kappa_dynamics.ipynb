## Linear rate network, $p=2$

### $\delta = 0$

We also check that for $\delta \approx 0$ it also works

```python
rnn = LowRankCyclingRNN.new_valentin(p=2, N=1_000_000, phi=identity, I_ext=zero, exclude_self_connections=True, delta=0, shift=1)
res = rnn.simulate(rnn.F[:, 0], (0, 40), dt_max=1, progress=True)
plot_neuron_trajectory(res);
plot_overlap_trajectory(rnn, res);
```

```python
rnn = LowRankCyclingRNN.new_valentin(p=2, N=1_000_000, phi=identity, I_ext=zero, exclude_self_connections=True, delta=0.1, shift=1)
res = rnn.simulate(rnn.F[:, 0], (0, 40), dt_max=1, progress=True)
plot_neuron_trajectory(res);
plot_overlap_trajectory(rnn, res);
```
